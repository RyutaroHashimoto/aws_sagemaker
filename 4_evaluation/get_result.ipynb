## Connect to AWS and get the results of training jobs that match the conditions.

### Search by tag name

```python
import boto3

search_params={
   "MaxResults": 10,
   "Resource": "TrainingJob",
   "SearchExpression": { 
      "Filters": [{ 
            "Name": "Tags.module",
            "Operator": "Equals",
            "Value": "Pytorch"
         },
          {
            "Name": "TrainingJobStatus",
            "Operator": "Equals",
            "Value": "Completed"
         }]},
#   "SortBy": "Metrics.train:binary_classification_accuracy",
#   "SortOrder": "Descending"
}

smclient = boto3.client(service_name='sagemaker')
results = smclient.search(**search_params)
```

```python
results
```

### Search by job name

```python
import boto3
search_params={
   "MaxResults": 100,
   "Resource": "TrainingJob",
   "SearchExpression": { 
      "Filters": [{ 
            "Name": "TrainingJobName",
            "Operator": "Contains",
            "Value": "tabnet"
         },
          {
            "Name": "TrainingJobStatus",
            "Operator": "Equals",
            "Value": "Completed"
         }]},
#   "SortBy": "Metrics.train:binary_classification_accuracy",
#   "SortOrder": "Descending"
}

smclient = boto3.client(service_name='sagemaker')
results = smclient.search(**search_params)
```

## convert to pd.DataFrame

```python
import pandas as pd
import numpy as np
t = pd.json_normalize(results['Results'])
met_list = [j['MetricName'] for j in t['TrainingJob.FinalMetricDataList'][0]] 
t[met_list] = np.nan
t["train_S3uri"] = np.nan
for i in range(len(t)):

    if len(t['TrainingJob.FinalMetricDataList'][i]) > 0:
        for j in range(len(t['TrainingJob.FinalMetricDataList'][i])):
            if len(t['TrainingJob.FinalMetricDataList'][i][j]) > 0:                 
                if  t['TrainingJob.FinalMetricDataList'][i][j]['MetricName'] in met_list:
                    t[t['TrainingJob.FinalMetricDataList'][i][j]['MetricName']][i] =  t['TrainingJob.FinalMetricDataList'][i][j]['Value']

                    
                    

    t['train_S3uri'][i] = t["TrainingJob.InputDataConfig"][i][0]['DataSource']['S3DataSource']['S3Uri']
```
